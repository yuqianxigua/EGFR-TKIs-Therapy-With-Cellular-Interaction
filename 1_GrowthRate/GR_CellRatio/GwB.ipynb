```python
## growth rate of W-Cell with ratio of X/W and Z/W (cell-to-cell interaction)
## osimertinib

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

# parameter
qyw = 0.08
pyw = 0.0009
qzw = 0.06
pzw = 0.0009
gw = list()
x = list()
y = list()

w1 = (qyw*0.00001)/(pyw+0.00001)+(qzw*0.00000001)/(pzw+0.00000001)-0.321
print (w1)

# x and y
for yw in np.arange(0, 0.01, 0.0001):
    for zw in np.arange(0, 0.00001, 0.0000001):
        x.append(yw)
        y.append(zw)
        gxzw = (qyw*yw)/(pyw+yw) + (qzw*zw)/(pzw+zw)-0.321
        gw.append(gxzw)

# fig
x = np.array(x)
print (len(x))
y = np.array(y)
print (len(y))
z = np.array(gw)
fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.plot_trisurf(x.flatten(), y.flatten(), z.flatten(), label = 'growth rate of W-cell')
# ax.set_zscale('log')
# ax.set_zlim(10**11, 10**13)
ax.set_xlabel('Y/W')
ax.set_ylabel('Z/W')
ax.set_zlabel('Growth rate of W') 
ax.set_title('drugB-Gw|total_cell#=10^9, Ymax=10^7, Zmax=10^6')

plt.show()
```

```python
## growth rate of W-Cell with ratio of X/W (cell-to-cell interaction)
## osimertinib

zw1 = 0.00000001
gw1 = list()
x1 = list()

# x
for yw1 in np.arange(0, 0.01, 0.0001):
    x1.append(yw1)
    gyzw1 = (qyw*yw1)/(pyw+yw1) + (qzw*zw1)/(pzw+zw1)-0.321
    gw1.append(gyzw1)

# fig
fig = plt.figure()
ax1 = fig.add_subplot()
plt.title('drugB-Gw|total_cell#=10^9, Xmax=10^7, Z=10')
plt.plot(x1, gw1)
plt.grid()
ax1.set_ylim((-0.4, -0.2))
plt.ylabel('growth rate of W-cell')
plt.xlabel('Y/W')

plt.show()
```
