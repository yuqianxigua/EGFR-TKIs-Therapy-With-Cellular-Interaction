```python
## growth rate of W-Cell with ratio of X/W and Z/W (cell-to-cell interaction)
## erlotinib

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

# parameter
qxw = 0.08
pxw = 0.0009
qzw = 0.06
pzw = 0.0009
gw = list()
x = list()
y = list()

w1 = (qxw*0.00001)/(pxw+0.00001)+(qzw*0.00000001)/(pzw+0.00000001)-0.17
print (w1)

# x and y
for xw in np.arange(0, 0.01, 0.0001):
    for zw in np.arange(0, 0.00001, 0.0000001):
        x.append(xw)
        y.append(zw)
        gxzw = (qxw*xw)/(pxw+xw) + (qzw*zw)/(pzw+zw)-0.17
        gw.append(gxzw)

# fig
x = np.array(x)
print (len(x))
y = np.array(y)
print (len(y))
z = np.array(gw)
fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.plot_trisurf(x.flatten(), y.flatten(), z.flatten(), label = 'growth rate of W-cell')
# ax.set_zscale('log')
# ax.set_zlim(10**11, 10**13)
ax.set_xlabel('Initial ratio of X')
ax.set_ylabel('Initial ratio of Z')
ax.set_zlabel('Growth rate of W') 
ax.set_title('drugAB-Gw|total_cell#=10^9, Xmax=10^7, Zmax=10^4')

plt.show()
```

```python
## growth rate of W-Cell with ratio of X/W (cell-to-cell interaction)
## erlotinib

zw1 = 0.00000001
gw1 = list()
x1 = list()

# x
for xw1 in np.arange(0, 0.01, 0.0001):
    x1.append(xw1)
    gxzw1 = (qxw*xw1)/(pxw+xw1) + (qzw*zw1)/(pzw+zw1)-0.081
    gw1.append(gxzw1)

# fig
fig = plt.figure()
ax1 = fig.add_subplot()
plt.title('drugAB-Gw|total_cell#=10^9, Xmax=10^7, Z=10, Y=10^4')
plt.plot(x1, gw1)
plt.grid()
ax1.set_ylim((-0.08, 0.02))
plt.ylabel('growth rate of W-cell')
plt.xlabel('Ratio of X-cell')

plt.show()
```
