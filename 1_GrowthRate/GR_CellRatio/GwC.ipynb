```python
## growth rate of W-Cell with ratio of Z/W (cell-to-cell interaction)
## combination

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

# parameter
qzw = 0.2
pzw = 0.0009
gw = list()
x = list()

x1 = (qzw*0.0000001)/(pzw+0.0000001)-0.064
print (x1)

# x and y
for zw in np.arange(0, 0.00001, 0.0000001):
    x.append(zw)
    gzw = ((qzw*zw)/(pzw+zw))-0.064
    gw.append(gzw)

# fig
fig = plt.figure()
ax = fig.add_subplot()
plt.title('drugC-Gw|Z/Wmax=0.00001')
plt.plot(x, gw)
ax.set_ylim(-0.065, -0.05)
plt.grid()
plt.ylabel('Growth Rate of W-cell')
plt.xlabel('Z/W(initial)')

plt.show()
```
