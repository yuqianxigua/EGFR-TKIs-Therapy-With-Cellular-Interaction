```python
## growth rate of Y-Cell with ratio of Z/Y (cell-to-cell interaction)
## osimertinib

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

# parameter
qzy = 0.007
pzy = 0.0009
gy = list()
x = list()

x1 = (qzy*0.001)/(pzy+0.001)+0.0203
print (x1)

# x and y
for zy in np.arange(0, 0.001, 0.000001):
    x.append(zy)
    gzy = ((qzy*zy)/(pzy+zy))+0.0203
    gy.append(gzy)

# fig
fig = plt.figure()
ax = fig.add_subplot()
plt.title('drugB-Gy|Z/Ymax=1')
plt.plot(x, gy)
ax.set_ylim((0.02, 0.03))
plt.grid()
plt.ylabel('Growth Rate of Y-cell')
plt.xlabel('Z/Y(initial)')

plt.show()
```
