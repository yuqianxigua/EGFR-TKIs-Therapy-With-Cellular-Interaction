```python
## growth rate of X-Cell with ratio of Z/X (cell-to-cell interaction)
## combination

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

# parameter
qzx = 0.025
pzx = 0.0009
gx = list()
x = list()

x1 = (qzx*0.000001)/(pzx+0.000001)-0.03355
print (x1)

# x and y
for zx in np.arange(0, 0.02, 0.00001):
    x.append(zx)
    gzx = ((qzx*zx)/(pzx+zx))-0.03355
    gx.append(gzx)

# fig
fig = plt.figure()
ax = fig.add_subplot()
plt.title('drugC-Gx|Z/Xmax=1')
plt.plot(x, gx)
# ax.set_ylim(-0.034, -0.01)
plt.grid()
plt.ylabel('Growth Rate of X-cell')
plt.xlabel('Z/X(initial)')

plt.show()
```
