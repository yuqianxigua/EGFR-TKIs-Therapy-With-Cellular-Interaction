```python
## growth rate of W-Cell with ratio of Z/X (cell-to-cell interaction)
## erlotinib

import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

# parameter
qzx = 0.07
pzx = 0.0009
gx = list()
x = list()

x1 = (qzx*0.001)/(pzx+0.001)+0.008
print (x1)

# x and y
for zx in np.arange(0, 0.01, 0.0001):
    x.append(zx)
    gzx = ((qzx*zx)/(pzx+zx))+0.008
    gx.append(gzx)

# fig
fig = plt.figure()
plt.title('drugAB-Gx|Z/Xmax=1')
plt.plot(x, gx)
plt.grid()
plt.ylabel('Growth Rate of X-cell')
plt.xlabel('Z/X(initial)')

plt.show()
```
